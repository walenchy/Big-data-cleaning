# STEP- 25 Unpivoting using pandas melt function which is quite crucialUnpivot data- i need to manipulate the social Deprivation table or Dataframe by unpivoting the columns into rowsmeaning if the some columns were on the column level, ineed to transform them into rows so as to be able to perform some intelligent ETL process like merging

```python
#first thing, so I will create a new variable ----> UnpivotedDomains_of_Deprivation

#after that I will tell python to keep the following columns as columns as it were ---> LSOA code', 'LSOA name(id_vars)'

#Then I will tell python to create a new column called category ----> var_name = 'category',
#and their row valuename should be --> value_name='value')

UnpivotedDomains_of_Deprivation = pd.melt(copy_Domains_of_deprivation, id_vars = ['LSOA code', 'LSOA name'], var_name= 'category', value_name='value')
```

```python
UnpivotedDomains_of_Deprivation.head()
```
